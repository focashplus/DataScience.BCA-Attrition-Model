```python
import matplotlib.pyplot as pyplot
from sklearn.metrics import roc_curve, roc_auc_score
def create_auc_chart(y,y_pred_prob):
        """
        Plot ROC curve
        """
# generate a no skill prediction (majority class)
no_skill_probs = [0 for _ in range(len(y))]

# calculate precision and recall for each threshold
fpr_base, tpr_base, _ = roc_curve(y, no_skill_probs)
fpr, tpr, thresholds_test = roc_curve(y, y_pred_prob)

pyplot.figure(figsize=(6,5))
        
# calculate score
ns_auc = roc_auc_score(y, no_skill_probs)
auc = roc_auc_score(y, y_pred_prob)

# summarize scores
print('Baseline: auc=%.3f' % (ns_auc))
print('Validation: auc=%.3f' % (auc))

# plot the precision-recall curves
pyplot.plot(fpr_base, tpr_base, linestyle='--')
pyplot.plot(fpr, tpr, linestyle='-', color = 'green', label='Validation AUC=%.3f' % auc)

# axis labels
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')
plt.plot([0, 1], [0, 1],'k--')
plt.xlim([-0.01, 1.01])
plt.ylim([-0.01, 1.01])
pyplot.title('ROC Curve')

# show the legend
pyplot.legend(loc = 'lower right')
```

```python
def compute_lift_chart(target_values, prob):


    pchurn    = pd.DataFrame({'Observed': target_values, 'p_churn': prob})
    sorted_df = pchurn.sort_values(by=['p_churn'], ascending=False)
    
    lift_chart         = np.zeros(10)
    cum_rel_lift_chart = np.zeros(10)
    
    dec_size       = int(np.ceil(sorted_df.shape[0]/10))
    overall_ratio  = sorted_df.Observed.sum()/sorted_df.shape[0]
    
    for i in range(0,10):
        lift_chart[i]         = 100*sorted_df.Observed.iloc[i*dec_size:(i+1)*dec_size].sum()/dec_size
        cum_rel_lift_chart[i] = sorted_df.Observed.iloc[:(i+1)*dec_size].sum()/((i+1)*dec_size)/overall_ratio
    
    return (np.arange(10,110,10), lift_chart, cum_rel_lift_chart)
```

```python
import matplotlib.pyplot as pyplot
def create_lift_chart(y,y_pred_train):
        """
        Plot lift curve
        """
        pyplot.figure(figsize=(8,6))

        x_train, _, lift_train = compute_lift_chart(y, y_pred_train)

        plt.plot(x_train, lift_train , color = 'green', label='Validation')
        plt.plot([10,100],[1,1],'r--')
        plt.xlabel('Cumulative population (%) ranked by p(Target3MDormancy)')
        plt.xticks(np.arange(10,110,10))
        plt.ylabel('Lift')
        plt.title('Lift chart')
        plt.legend(loc=1)
        plt.show()

       # print('Test Set Lift:')
        print(lift_train)
        #print(cum_rel_lift_chart)
```

```python
X = pd.merge(X,df[['ACCNO']] ,how ='left',left_index = True, right_index = True)
X = pd.merge(X,df[['TransactionMonth']] ,how ='left',left_index = True, right_index = True)
```

```python
#The '.predict' and 'predict_proba' doesn't change the order of classified cases
df_validation = pd.concat([X.reset_index(drop=True),y.reset_index(drop=True),y_pred.reset_index(drop=True),y_pred_prob.reset_index(drop='True')],axis=1)
```

```python
df_validation.head()
```

```python
df_validation.rename(columns = {1:'Probability' ,
                          0:'Prediction'},  inplace = True)
```

```python
 def Binned_prob(Probability):
    if Probability==0:
        return '0'
    elif Probability<=0.1:
        return '(0-0.1]'
    elif Probability<=0.2:
        return '(0.1-0.2]'
    elif Probability<=0.3:
        return '(0.2-0.3]'
    elif Probability<=0.4:
        return '(0.3-0.4]'
    elif Probability<=0.5:
        return '(0.4-0.5]'
    elif Probability<=0.6:
        return '(0.5-0.6]'
    elif Probability<=0.7:
        return '(0.6-0.7]'
    elif Probability<=0.8:
        return '(0.7-0.8]'
    elif Probability<=0.9:
        return '(0.8-0.9]'
    elif Probability<=1:
        return '(0.9-1]'  
    else:
        return 'Null'

df_validation['Binned_Probability']= df_validation['Probability'].apply(Binned_prob)
```
