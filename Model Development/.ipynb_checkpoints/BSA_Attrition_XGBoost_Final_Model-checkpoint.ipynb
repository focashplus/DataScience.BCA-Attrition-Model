### 2.Feature Elimination with Reduce_Vars Function

```python
import joblib 
from reduce_vars import reduce_variables
```

```python
def reduce_variables(model_object,pandas_dataset,corr_limit=0.50):
    ''' Produces an optimized variable list based on a classifier variable importance and variable correlation. Choosing a list of un-corrlated variables with an importance priority. Returns a list of variable names. 
    Attributes:
    model_object: classifier object such as a sklearn random forest or any other that has feature_importances_ and feature_names_in_ attributes
    pandas_dataset: a pandas dataframe that the classifier object was build on
    corr_limit: a correlation limit 0.5 by default  '''
    import pandas as pd

    # adding variable importance 
    variable_importance = pd.DataFrame(list(zip(model_object.get_booster().feature_names,model_object.feature_importances_)))
    #variable_importance = pd.DataFrame(list(zip(model_object.feature_names_in_,model_object.feature_importances_)))
    variable_importance.columns = ['Name','Importance']
    variable_importance.sort_values(by='Importance',ascending=False,inplace=True)

    # removing variables with an importance of 0
    drop_vars = variable_importance['Name'].loc[variable_importance['Importance'] == 0]
    drop_vars = list(drop_vars)
    # drop_vars 
    pandas_dataset.drop(columns=drop_vars,inplace=True)

    # Correlation
    
    corr = pandas_dataset.corr().stack().reset_index()
    corr.columns = ['var1','var2','corr']
    # creating an abs corr value column 
    corr['abs_corr'] = abs(corr['corr'])
    
    # looping and creting a keep list
    vars_to_keep = []
    important_vars = variable_importance.loc[variable_importance['Importance'] != 0]

    while True:
        if len(important_vars['Name']) == 0:
            print('All variables accounted for')
            break
        else:
            important_vars = important_vars.sort_values(by='Importance',ascending=False)
            important_vars.reset_index(inplace=True)
            important_vars.drop(columns='index',inplace=True)
            most_important_var_loop = important_vars.iloc[0][0] # taking first var from the top
            vars_to_keep.append(most_important_var_loop)
            exclude_vars = list(corr['var2'].loc[(corr['var1'] == most_important_var_loop)&(corr['abs_corr'] >= corr_limit)])
            important_vars = important_vars.loc[~important_vars['Name'].isin(exclude_vars)]

    return vars_to_keep
```

```python
def compute_lift_chart(target_values, prob):


    pchurn    = pd.DataFrame({'Observed': target_values, 'p_churn': prob})
    sorted_df = pchurn.sort_values(by=['p_churn'], ascending=False)
    
    lift_chart         = np.zeros(10)
    cum_rel_lift_chart = np.zeros(10)
    
    dec_size       = int(np.ceil(sorted_df.shape[0]/10))
    overall_ratio  = sorted_df.Observed.sum()/sorted_df.shape[0]
    
    for i in range(0,10):
        lift_chart[i]         = 100*sorted_df.Observed.iloc[i*dec_size:(i+1)*dec_size].sum()/dec_size
        cum_rel_lift_chart[i] = sorted_df.Observed.iloc[:(i+1)*dec_size].sum()/((i+1)*dec_size)/overall_ratio
    
    return (np.arange(10,110,10), lift_chart, cum_rel_lift_chart)
```

```python
import matplotlib.pyplot as pyplot
def create_lift_chart(y_train,y_pred_prob_train,y_test,y_pred_prob_test):
        """
        Plot lift curve
        """
        pyplot.figure(figsize=(8,6))
        x, _, lift_test = compute_lift_chart(y_test,y_pred_prob_test)
        x_train, _, lift_train = compute_lift_chart(y_train, y_pred_prob_train)

        plt.plot(x, lift_test , color = 'red', label='Test')
        plt.plot(x_train, lift_train , color = 'blue', label='Train')
        plt.plot([10,100],[1,1],'r--')
        plt.xlabel('Cumulative population (%) ranked by p(churn)')
        plt.xticks(np.arange(10,110,10))
        plt.ylabel('Lift')
        plt.title('Lift chart')
        plt.legend(loc=1)
        plt.show()

        print('Test Set Lift:')
        print(lift_test)
        #print(cum_rel_lift_chart)
```

```python
def compute_lift_chart(target_values, prob):


    pchurn    = pd.DataFrame({'Observed': target_values, 'p_churn': prob})
    sorted_df = pchurn.sort_values(by=['p_churn'], ascending=False)
    
    lift_chart         = np.zeros(10)
    cum_rel_lift_chart = np.zeros(10)
    
    dec_size       = int(np.ceil(sorted_df.shape[0]/10))
    overall_ratio  = sorted_df.Observed.sum()/sorted_df.shape[0]
    
    for i in range(0,10):
        lift_chart[i]         = 100*sorted_df.Observed.iloc[i*dec_size:(i+1)*dec_size].sum()/dec_size
        cum_rel_lift_chart[i] = sorted_df.Observed.iloc[:(i+1)*dec_size].sum()/((i+1)*dec_size)/overall_ratio
    
    return (np.arange(10,110,10), lift_chart, cum_rel_lift_chart)
```

```python
import matplotlib.pyplot as pyplot
def create_lift_chart(y_train,y_pred_prob_train,y_test,y_pred_prob_test):
        """
        Plot lift curve
        """
        pyplot.figure(figsize=(8,6))
        x, _, lift_test = compute_lift_chart(y_test,y_pred_prob_test)
        x_train, _, lift_train = compute_lift_chart(y_train, y_pred_prob_train)

        plt.plot(x, lift_test , color = 'red', label='Test')
        plt.plot(x_train, lift_train , color = 'blue', label='Train')
        plt.plot([10,100],[1,1],'r--')
        plt.xlabel('Cumulative population (%) ranked by p(churn)')
        plt.xticks(np.arange(10,110,10))
        plt.ylabel('Lift')
        plt.title('Lift chart')
        plt.legend(loc=1)
        plt.show()

        print('Test Set Lift:')
        print(lift_test)
        print(lift_train)
        #print(cum_rel_lift_chart)
```
